```python
import os

# Section 1.

banner = "=" * 42 #string repitition banner to welcome
project_name = input("Enter project name: ")
project_version = input("Enter project version: ")
year_started = int(input("Enter year started: "))
main_language = input("Enter main language: ")
project_lead = input("Enter project lead name: ")

project = (project_name, project_version, year_started, main_language, project_lead)

print(banner)
print("Welcome!")
print(banner)
print(f" {project[0]} — Open Source Project")
print(f" Project Lead : {project[4]}")
print(banner)

print(f"project[0] : {project[0]}")
print(f"project[1] : {project[1]}")
print(f"project[2] : {project[2]}")
print(f"project[3] : {project[3]}")
print(f"project[4] : {project[4]}")
print(f"First 3 fields : {project[:3]}")
print(f"Language count : {project.count(project[3])} Language index : {project.index(project[3])}")

# project[0] = 'test'  # TypeError: tuple does not support item assignment
# Tuples suit this data because official project details should not be changed by mistake.

contributors = []

for i in range(4):
    print(f"\nEnter contributor {i + 1} details")
    contributor_name = input("Name: ")
    contributor_role = input("Role: ")
    contributor_language = input("Language: ")
    contributor_commits = int(input("Commits: "))
    contributor_country = input("Country: ")

    contributor = {
        'name': contributor_name,
        'role': contributor_role,
        'language': contributor_language,
        'commits': contributor_commits,
        'country': contributor_country
    }
    contributors.append(contributor)

names = []
for contributor in contributors:
    names.append(contributor['name'])

names.sort()
print("-" * 42)
print(f"Sorted names : {names}")
print(f"Last name : {names[-1]} First two : {names[:2]}")

for contributor in contributors:
    contributor.update({'status': 'Active'})

print(f"Contributor 1 status : {contributors[0].get('status')}")
backup_contributor = contributors[0].copy()
print(f"Backup : {backup_contributor}")

# ==============================
# Section 2.
# ==============================

issues = []

for i in range(5):
    print(f"\nEnter issue {i + 1} details")
    issue_id = input("Issue id: ")
    issue_title = input("Title: ")
    issue_type = input("Type (Bug/Feature): ")
    issue_priority = input("Priority (Critical/High/Medium/Low): ")
    issue_reporter = input("Reporter: ")
    issue_status = input("Status (Open/In Progress/Resolved): ")

    issue = {
        'id': issue_id,
        'title': issue_title,
        'type': issue_type,
        'priority': issue_priority,
        'reporter': issue_reporter,
        'status': issue_status
    }
    issues.append(issue)

open_count = 0
for issue in issues:
    if issue['status'] == 'Open':
        open_count += 1

print("-" * 42)
print(f"Open issues : {open_count}")

issues[0]['priority'] = 'Critical'
print("First issue → priority updated to Critical.")
print(f"Last two issues : {issues[-2:]}")

reporters = set()
for issue in issues:
    reporters.add(issue['reporter'])

tech_stack = set()
for contributor in contributors:
    tech_stack.add(contributor['language'])

tech_stack.add('Git')
tech_stack.discard('PHP')

union_set = reporters.union(tech_stack)
intersection_set = reporters.intersection(tech_stack)
difference_set = reporters.difference(tech_stack)

print("-" * 42)
print(f"reporters : {reporters}")
print(f"tech_stack : {tech_stack}")
print(f"union : {union_set}")
print(f"intersection : {intersection_set} difference : {difference_set}")

priority_set = set()
for issue in issues:
    priority_set.add(issue['priority'])

if 'Critical' in priority_set:
    print("Critical present : YES — flag for immediate review.")
else:
    print("Critical present : NO — no critical issue right now.")

priority_count = {}
for issue in issues:
    priority = issue['priority']
    if priority in priority_count:
        priority_count[priority] += 1
    else:
        priority_count[priority] = 1

status_groups = {}
for issue in issues:
    status = issue['status']
    if status not in status_groups:
        status_groups[status] = []
    status_groups[status].append(issue['title'])

print("-" * 42)
print(f"priority keys : {priority_count.keys()}")
print(f"priority values : {priority_count.values()}")
print(f"priority items : {priority_count.items()}")

for status in status_groups.keys():
    print(f"Status Groups → {status} : {status_groups[status]}")

reporter_count = {}
for issue in issues:
    reporter = issue['reporter']
    if reporter in reporter_count:
        reporter_count[reporter] += 1
    else:
        reporter_count[reporter] = 1

top_reporter = ''
top_reporter_count = 0
for reporter in reporter_count:
    if reporter_count[reporter] > top_reporter_count:
        top_reporter = reporter
        top_reporter_count = reporter_count[reporter]

print(f"Top reporter : {top_reporter} ({top_reporter_count} issues)")

issues[0].pop('type')
print(f"After pop('type') : {issues[0]}")

# ==============================
# Section 3.
# ==============================

folder_name = project[0].lower().replace(' ', '_')

if os.path.exists(folder_name):
    print(f"Folder already exists : {folder_name}/")
else:
    os.mkdir(folder_name)
    print(f"Folder created : {folder_name}/")

report_path = os.path.join(folder_name, 'project_report.txt')
csv_path = os.path.join(folder_name, 'issues.csv')

priority_summary_parts = []
for priority in priority_count:
    priority_summary_parts.append(f"{priority}:{priority_count[priority]}")
priority_summary = ' '.join(priority_summary_parts)

report_lines = []
report_lines.append("=" * 42)
report_lines.append(f"{project[0]} — PROJECT REPORT")
report_lines.append("=" * 42)
report_lines.append(f"Project Name : {project[0]}")
report_lines.append(f"Version : {project[1]}")
report_lines.append(f"Year Started : {project[2]}")
report_lines.append(f"Main Language : {project[3]}")
report_lines.append(f"Lead : {project[4]}")
report_lines.append("")
report_lines.append("Contributors")
for contributor in contributors:
    report_lines.append(f"{contributor['name']} | {contributor['role']} | {contributor['language']} | {contributor['commits']} | {contributor['country']} | {contributor['status']}")
report_lines.append("")
report_lines.append("Issues")
for issue in issues:
    issue_type_text = issue.get('type', 'Removed')
    report_lines.append(f"{issue['id']} | {issue['title']} | {issue_type_text} | {issue['priority']} | {issue['reporter']} | {issue['status']}")
report_lines.append("")
report_lines.append("Analysis")
report_lines.append(f"Open Issues : {open_count}")
report_lines.append(f"Reporters : {reporters}")
report_lines.append(f"Tech Stack : {tech_stack}")
report_lines.append(f"Priority Breakdown : {priority_summary}")
report_lines.append(f"Top Reporter : {top_reporter} ({top_reporter_count} issues)")

try:
    with open(report_path, 'w', encoding='utf-8') as file:
        for line in report_lines:
            file.write(f"{line}\n")

    with open(csv_path, 'w', encoding='utf-8') as file:
        file.write("id,title,priority,reporter,status\n")
        for issue in issues:
            file.write(f"{issue['id']},{issue['title']},{issue['priority']},{issue['reporter']},{issue['status']}\n")

    print(f"Files saved : {os.listdir(folder_name)}")
except IOError:
    print("File writing error happened.")

try:
    print("--- read() ---")
    with open(report_path, 'r', encoding='utf-8') as file:
        print(file.read())

    print("--- readline() ---")
    with open(report_path, 'r', encoding='utf-8') as file:
        print(f"Line 1 : {file.readline().strip()}")
        print(f"Line 2 : {file.readline().strip()}")

    print("--- readlines() ---")
    with open(report_path, 'r', encoding='utf-8') as file:
        all_lines = file.readlines()

    important_lines = []
    for line in all_lines:
        if 'Critical' in line or 'High' in line:
            important_lines.append(line.strip())

    print(f"Total lines : {len(all_lines)} Critical/High lines : {len(important_lines)}")
    for line in important_lines:
        print(line)
except FileNotFoundError:
    print("File not found while reading.")

# ==============================
# Bonus.
# ==============================

urgent = [issue['title'] for issue in issues if issue['priority'] == 'Critical' or issue['priority'] == 'High']
print("-" * 42)
print(f"Urgent issues : {urgent}")
print(f"Urgent issues count : {len(urgent)}")

try:
    with open(report_path, 'a', encoding='utf-8') as file:
        file.write("\nURGENT ISSUES\n")
        for title in urgent:
            file.write(f"- {title}\n")

    with open(report_path, 'r', encoding='utf-8') as file:
        final_lines = file.readlines()

    print("--- Last 6 lines after append ---")
    for line in final_lines[-6:]:
        print(line.strip())
    print("Append confirmed.")
except FileNotFoundError:
    print("Report file not found for append check.")
except IOError:
    print("File append error happened.")

# ==============================
# Final Summary
# ==============================

print("=" * 42)
print(f" {project[0]} — FINAL SUMMARY")
print("=" * 42)
print(f"Project : {project[0]} Version : {project[1]} Lead : {project[4]}")
print(f"Contributors : {len(contributors)} Names : {names}")
print(f"Tech Stack : {tech_stack}")
print(f"Issues : {len(issues)} Open : {open_count} Reporters : {reporters}")
print(f"Top Reporter : {top_reporter} ({top_reporter_count} issues)")
print(f"Priority Breakdown : {priority_summary}")
print(f"Report : {report_path}")
print(f"CSV : {csv_path}")
print("=" * 42)
print(f"{project[0]} complete. Thank you for contributing to open source!")
print("=" * 42)
```
